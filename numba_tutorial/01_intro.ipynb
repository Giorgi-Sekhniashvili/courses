```python
from numba import jit, vectorize
import random
```

```python
def monte_carlo_pi(n_samples):
    acc = 0
    for i in range(n_samples):
        x = random.random()
        y = random.random()

        if (x**2 + y**2) < 1.0:
            acc += 1
    return 4.0 * acc / n_samples
```

```python
%time monte_carlo_pi(10_000_000)
```

```python
# jit(nopython = True): if the types of elements are different this will give an error
monte_carlo_pi = jit()(monte_carlo_pi)
```

```python
%time monte_carlo_pi(10_000_000)
```

### what you should avoid!!!
1. don't use different type of elements in lists. __njit()__ makes shure that it won't happen
2. don't use lists, use np.arrays
3. __@vectorize__ is fast decorator too. This works on a single values as well as the lists


```python
@vectorize
def scalar_computation(num):
    if num % 2 == 0:
        return 2
    else:
        return 1

scalar_computation(15)
```

```python
def scalar_computation_on_lists(list):
    output = []
    for item in list:
        if item % 2 == 0:
            output.append(2)
        else:
            output.append(1)
    return output
```

```python
test_array = list(range(100_000))
```

```python
%time scalar_computation_on_lists(test_array)
```

```python
%time scalar_computation(test_array)
```
