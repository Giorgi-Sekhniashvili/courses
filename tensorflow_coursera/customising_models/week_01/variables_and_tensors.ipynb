```python
import tensorflow as tf
import numpy as  np
tf.__version__
```

```python
my_var = tf.Variable([1, 3, 5], dtype=tf.float32)
my_var
```

```python
print(my_var.numpy())
my_var.assign_add([16, 16, 16])

print(my_var.numpy())
```

```python
strings = tf.Variable(['Hola Hola'], dtype=tf.string)
ints = tf.Variable([2, 4, 5], dtype=tf.int32)
complexes = tf.Variable([23 - 1j, 23 - 2j], dtype=tf.complex64)

strings, ints, complexes
```
