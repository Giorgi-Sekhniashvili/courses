```python
import tensorflow as tf
import numpy as np

tf.__version__
```

```python
vgg_model = tf.keras.applications.VGG16()
vgg_model.summary()
```

```python
vgg_model.trainable = False
out = vgg_model.get_layer('flatten').output
out = tf.keras.layers.Dense(10, activation='relu', trainable=False, name='dense_1')(out)
out = tf.keras.layers.Dense(10, activation='sigmoid', name='dense_2')(out)
my_model = tf.keras.Model(inputs=vgg_model.inputs, outputs=out)
my_model.summary()
```
